```python
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
df_train = pivot_df.head(45)
df_valid = pivot_df[45:]
```

```python
# 從 df_train 中分離出被解釋變數 y
growth_train = df_train['NE0904']
df_train_X = df_train.drop(columns=['NE0904'])

growth_valid = df_valid['NE0904'] 
df_valid_X = df_valid.drop(columns=['NE0904'])
 
Lasso_l = Pipeline(steps=[('poly', PolynomialFeatures(degree=1)), ('Lasso', Lasso(alpha=1000))])

large = Lasso_l.fit(df_train_X, growth_train)
growth_pred_l = large.predict(df_valid_X)
large_alpha = list(growth_pred_l)
```
